```python
# 格式标签列表生成器-GUI
import tkinter as tk
from tkinter import messagebox

def vList(temp:list, p):
    '''【已弃用】新建按钮递归函数'''
    deth = 3
    for elem in temp:
        if type(elem) == list:
            vList(elem, p)
        else:
            print(elem)
            tempbtn = tk.Button(p, text = str(elem), width=50)
            tempbtn.grid(row = depth, column = 1)
            depth = depth + 1

def showChoices(choices:list, btns):
    if choices.__len__() > 4:
        flag = 4
        for choice in choices:
            if not flag:break
    else:
        for i in range(0, choices.__len__()):
            btns[i]['text'] = choices[i]

def updateBtn(btns):
    for btn in btns:
        btn.grid_forget()

def up(btn_text, Text):
    Text.insert("insert", btn_text)

def chooseOneTab(Tabs:list):
    '''【此处仅完成了GUI设计】基于TabTree生成一个长标签并返回'''
    root = tk.Tk()
    root.title('格式标签列表生成器')
    root.geometry('370x500+1000+50')
    tempTab = tk.Text(root, width=50, height=3)
    tempTab.grid(row=1, column=1, columnspan=2)
    btn_withdraw = tk.Button(root, text='撤回', width=25)
    btn_withdraw.grid(row=2, column=1)
    btn_reverse_withdraw = tk.Button(root, text='反撤回', width=25)
    btn_reverse_withdraw.grid(row=2, column=2)
    # templist = [[1, [11, [21, 22], 12]], [2, [21, 22]]]
    # print(templist)
    # vList(templist, root)
    btns = []
    for ch in 'ABCD':
        btns.append(tk.Button(root, text=ch + str(Tabs['ABCD'.index(ch)]), width=50,
                    anchor='w', command=lambda: up(ch, tempTab)))
        btns['ABCD'.index(ch)].grid(row='ABCD'.index(ch) + 4, column=1, columnspan=2)
    btn_sure = tk.Button(root, text='确认', width = 50, command=lambda:updateBtn(btns))
    btn_sure.grid(row = 3, column = 1, columnspan = 2)
    showChoices(Tabs, btns)
    Tab = tempTab.get('0.0', tk.END)
    root.mainloop()
    return Tab

chooseOneTab([1, 2, 3, 4])
```

```python
# 函数部分
from xml.dom.minidom import parse
from tkinter import simpledialog
import tkinter as tk
import os
import re
import urllib.request
import sqlite3

def get_title(url):
    '''读取网站链接的标题'''
    try:
        page = urllib.request.urlopen(url=url)
    except Exception as e:
        return str(e)
    else:
        html = page.read().decode('utf-8')
        title = re.findall('<title>(.+)</title>', html)
        return title

def askTab(TabTree, Element, root, add_btn_row, Text, XmlFileName):
    result = tk.simpledialog.askstring(title='询问信息', prompt='请输入新标签：', initialvalue='Tab')
    if result == None or result == '':return 0
    # 向xml指定结点下，写入tab标签
    # 需要参数：XmlFileName, Element
    tab_node = TabTree.createElement("tab")
    tab_node.setAttribute("text", result)
    Element.appendChild(tab_node)
    with open(XmlFileName, 'w', encoding='utf-8') as f:
        TabTree.writexml(f, addindent='', encoding='utf-8')

    # 在加号下面，新建tab按钮，如加号下面有之前的按钮，则考虑直接覆盖
    # 于是乎，还是选择使用Label，按钮要传很多参数，而且本身Text也能被直接修改……
    # 需要参数：root, add_btn_row
    # 【弃用】add_btn = tk.Button(root, text = result, width=50)
    # 【弃用】add_btn.grid(row = add_btn_row, column = 1, columnspan=2)
    add_label_row = add_btn_row
    tk.Label(root, text = '成功新增标签：“' + result + '”', width=50).grid(row=add_label_row, column=1, columnspan=2)
    
    # 同时向Text追加输入标签
    Text.insert("insert", '\\'+result)
def child_text_list(childNodes):
    '''读取childNodes列表，返回其中tab结点的text属性列表及其各个元素对应的位置数列表'''
    l_text = []
    l_childNum = []
    num = 0
    for child in childNodes:
        if child.nodeName == "tab":
            l_text.append(child.getAttribute("text"))
            l_childNum.append(num)
        num += 1
    return l_text, l_childNum
def nextTabs(childNodes, tab):
    '''调用child_text_list函数，返回childNodes中text属性为tab的child_text_list返回值'''
    l1, l2 = child_text_list(childNodes)
    return child_text_list(childNodes[l2[l1.index(tab)]].childNodes)
def updateBtns(TabTree, childNodes, l_text, l_num, btns, root, base_row, one_choose, showText):
    '''LMQ: 参数还挺多，有9个，其实可以用列表输入，但其实无所谓，这个函数是专为chooseTabs方法中的活动控件代码段写的，不被其他函数调用'''
    for btn in btns:
        # 销毁原有按钮组
        btn.destroy()
    btns = []
    Element = childNodes[one_choose.get()]
    if l_text.__len__() == 0:
        tk.Label(root, text='已经到底啦~~', width=50).grid(row=base_row, column=1, columnspan=2)
        a = tk.Button(root, text='+', width=50, command=lambda: askTab(
            TabTree, Element, root, base_row + 2, showText, XmlFileName))
        a.grid(row=1 + base_row, column=1, columnspan=2)
        btns.append(a)
    else:
        # print(TabTree, childNodes, l_text, l_num, btns, root, base_row, one_choose.get(), showText)
        # print(one_choose.get()) # 参数多了，我自己都记不太清了，不懂就print。打印一下，一目了然！
        # 【此递进逻辑中被修复的Bug】Element = childNodes[l_num[one_choose.get()]]
        childNodes = childNodes[one_choose.get()].childNodes # 形参的“艺术”
        for text in l_text:
            r = tk.Radiobutton(root, text=text, width=50, variable=one_choose, 
                        value=l_num[l_text.index(text)], anchor='w', 
                        command=lambda: (upload(l_text[l_num.index(one_choose.get())], showText)
                               , updateBtns(TabTree, childNodes,
                            nextTabs(childNodes, l_text[l_num.index(one_choose.get())])[0],
                            nextTabs(childNodes, l_text[l_num.index(one_choose.get())])[1], 
                            btns, root, base_row, one_choose, showText)))
            r.grid(row=l_text.index(text) + 1 + base_row, column=1, columnspan=2)
            btns.append(r)
            if l_text.index(text) + 1 == l_text.__len__():
                a = tk.Button(root, text='+', width=50, command=lambda: askTab(
                    TabTree, Element, root, base_row+btns.__len__()+1, showText, XmlFileName))
                a.grid(row=l_text.index(text) + 2 + base_row, column=1, columnspan=2)
                btns.append(a)
def upload(btn_text, Text):
    Text.insert("insert", '\\'+btn_text)
def read_Text(Text, Tabs, root):
    r_tab = ''
    flag = 2
    for ch in Text.get(1.0, 'end'):
        if flag == 2 and ch == '\\':
            flag = 1
        elif flag == 1 and ch == '\\':
            flag = 0
            r_tab += ':'
        else:
            if not ch == '\n':r_tab += ch
    flag = 2
    Tabs.append(r_tab)
    Text.delete(1.0, 'end')
    root.destroy()
def change_tab_res():
    '''未完成'''
    print('更换资源中（待完善相关功能）')
def chooseTab(XmlFileName):
    '''XmlFileName为标签来源。\n
        返回一个空字符串或用户选择的标签。\n
        心有余而力不足，目前只做到了按一次确定生成一个长标签。'''
    root = tk.Tk()
    root.title('格式标签列表生成器')
    root.geometry('370x500+1000+50')
    Tabs = []

    # 基础控件
    showText_row = 3
    fileLable = tk.Label(root, text = '标签来源：'+XmlFileName, anchor = 'e', width=50)
    fileLable.grid(row=showText_row-2, column=1, columnspan=2)
    btn_xml = tk.Button(root, text='更换标签资源', width=50, command=lambda:change_tab_res())
    btn_xml.grid(row=showText_row-1, column=1, columnspan=2)
    showText = tk.Text(root, width=50, height=3)
    showText.grid(row=showText_row, column=1, columnspan=2)
    btn_withdraw = tk.Button(root, text='撤回【暂时不可用】', width=24) # 类似出栈
    btn_withdraw.grid(row=showText_row+1, column=1)
    btn_reverse_withdraw = tk.Button(
        root, text='反撤回【暂时不可用】', width=24)  # 类似入栈
    btn_reverse_withdraw.grid(row=showText_row+1, column=2)

    # 活动控件
    base_row = showText_row + 3
    TabTree = parse(XmlFileName)
    rootNode = TabTree.documentElement
    l_text, l_num = child_text_list(rootNode.childNodes)
    print(l_text, l_num)
    one_choose = tk.IntVar()
    # print(one_choose.get())
    btns = []
    for text in l_text:
        # LMQ：这段for循环代码，我研究了好久！涉及一个“按钮终值问题”。
        # 最终我使用“伴生列表算法”，在必须使用不定数循环、不允许新建新按钮类的条件下，独立解决了这个问题！
        r = tk.Radiobutton(root, text=text, width=50, variable=one_choose, 
                    value=l_num[l_text.index(text)], anchor='w', 
                           command=lambda: (upload(l_text[l_num.index(one_choose.get())], showText)
                           , updateBtns(TabTree, rootNode.childNodes,
                           nextTabs(rootNode.childNodes, l_text[l_num.index(one_choose.get())])[0], 
                           nextTabs(rootNode.childNodes, l_text[l_num.index(one_choose.get())])[1], 
                           btns, root, base_row, one_choose, showText)))
        r.grid(row=l_text.index(text) + 1 + base_row, column=1, columnspan=2)
        btns.append(r)
        if l_text.index(text) + 1 == l_text.__len__():
            a = tk.Button(root, text='+', width=50,
                          command=lambda: askTab(TabTree, rootNode, root, base_row+btns.__len__()+1, showText, XmlFileName))
            a.grid(row=l_text.index(text) + 2 + base_row, column=1, columnspan=2)
            btns.append(a)
    btn_sure = tk.Button(root, text='确认(标签框内为空时按下以结束输入)', width=50, command=lambda: read_Text(showText, Tabs, root)
                        #    | updateBtns(rootNode.childNodes, 
                        #    nextTabs(rootNode.childNodes, l_text[l_num.index(one_choose.get())])[0], 
                        #    nextTabs(rootNode.childNodes, l_text[l_num.index(one_choose.get())])[1], 
                        #    btns, root, base_row, one_choose, showText)
                           )
    btn_sure.grid(row=showText_row + 2, column=1, columnspan=2)

    root.resizable(height=False, width=False)
    root.mainloop()

    if Tabs.__len__() == 0:return ''
    return Tabs[0]
def get_Tabs_base_on_xml(XmlFileName):
    Tabs = []
    tab = ' '
    while True:
        tab = chooseTab(XmlFileName)
        if tab == '' : return Tabs
        Tabs.append(tab)

XmlFileName = r'E:\RER\3_Reference\《执行！执行！执行！》\TabTree.xml'

def db_add_link(XmlFileName):
    link_win = tk.Tk()
    link_win.title('新增网站链接')
    link_win.geometry('400x400+50+50')

    tk.Button(link_win, text = '+link', width=20, command = lambda: get_Tabs_base_on_xml(XmlFileName)).grid(row=1, column=2)
    tk.Button(link_win, text='+title', width=20,
              command=lambda: get_Tabs_base_on_xml(XmlFileName)).grid(row=2, column=2)
    tk.Button(link_win, text='+tabs', width=20,
              command=lambda: get_Tabs_base_on_xml(XmlFileName)).grid(row=3, column=2)

    link_win.mainloop()
get_Tabs_base_on_xml(XmlFileName)
# db_add_link(XmlFileName)
```
